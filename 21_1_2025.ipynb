```
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'species': ['setosa', 'versicolor', 'virginica', 'setosa', 'versicolor',
                'setosa', 'virginica', 'virginica', 'setosa', 'versicolor']
}
df = pd.DataFrame(data)


species_count = df['species'].value_counts()


species_count.plot(kind='bar' , color='black')


plt.title('Species Count in Dataset')
plt.xlabel('Species')
plt.ylabel('Count')


plt.show()
```
